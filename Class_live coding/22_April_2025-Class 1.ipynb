```python
np.random.seed(22)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x,y)
ax.set_title ('total growth over time')
ax.set_ylabel ('total growth')
ax.set_xlabel ('years since start')
fig.tight_layout ()
```

```python
#font dictionary
font1 = {'family': 'times new roman', 'color':'purple', 'size' : 18}
font2 = {'family': 'monospace', 'color':'cyan', 'size' : 18}
```

```python
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x,y)
ax.set_title ('total growth over time',fontdict= font1)
ax.set_ylabel ('total growth', fontdict= font2)
ax.set_xlabel ('years since start', fontdict= font2)
fig.tight_layout ()
```

```python
#change alignment of label 
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x,y)
ax.set_title ('total growth over time',fontdict= font1, loc= 'right')
ax.set_ylabel ('total growth', fontdict= font2)
ax.set_xlabel ('years since start', fontdict= font2)
fig.tight_layout ()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
              y,
              marker='H',         # NEW ADDITIONS
              color = "indigo")   # NEW ADDITIONS
```

```python
#modifying data points 
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter (x,
                y,
                marker = '*',
                color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,
        y,
        marker='H', 
        color = 'indigo', 
        linestyle = '--',   # NEW ADDITIONS
        linewidth = 2)      # NEW ADDITIONS
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
        markersize = 12,              # NEW ADDITIONS
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',  # NEW ADDITIONS
        markerfacecolor = '#fa9359' ) # NEW ADDITIONS
```

```python
#adding grid lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
        markersize = 12,              # NEW ADDITIONS
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',  # NEW ADDITIONS
        markerfacecolor = '#fa9359' ) # NEW ADDITIONS

ax.grid (axis = 'y', color ='blue',linewidth = 2, linestyle = '-.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(1)
x1 = np.arange(51)
y1 = np.random.randint(0,75,51)# random integers 0 to 75, 51 numbers
x2 = np.array(["chicken","duck","dog","rat","rabbit"])
y2 = np.array([110,180,240,99,220])
```

```python
#we want two graphs side by side
fig, (ax1,ax2) = plt.subplots (ncols = 2,
                               nrows =1,
                               figsize = (7,3))
ax1.scatter(x1,y1)
ax2.bar (x2,y2)
```

```python

# adding Axis Title
fig, (ax1,ax2) = plt.subplots (ncols = 2,
                               nrows =1,
                               figsize = (10,3))
ax1.scatter(x1,y1)
ax2.bar (x2,y2)

font1 = { "family":"serif", "color":"indigo"}

ax1.set_xlabel ("Time (hours)", fontsize = 12, fontdict = font1)
ax1.set_ylabel ("Number", fontsize = 12, fontdict = font1)
ax2.set_xlabel ("Farm Animal", fontsize = 12, fontdict = font1)
ax2.set_ylabel ("Quantity", fontsize = 12, fontdict = font1)

fig.show()
```

```python
# changing colors
fig, (ax1,ax2) = plt.subplots (ncols = 2,
                               nrows =1,
                               figsize = (10,3))
ax1.scatter(x1,y1, color = "cyan")
ax2.bar (x2,y2, color = "yellow")

font1 = { "family":"serif", "color":"black"}

ax1.set_xlabel ("Time (hours)", fontsize = 12, fontdict = font1)
ax1.set_ylabel ("Volume of water (mL)", fontsize = 12, fontdict = font1)
ax2.set_xlabel ("Farm Animal", fontsize = 12, fontdict = font1)
ax2.set_ylabel ("Quantity", fontsize = 12, fontdict = font1)
fig.show()
```

```python
#plotting 2 types on graphs on 1 plot
x = np.array(["chicken","duck","dog","rat","rabbit"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax =plt.subplots (figsize = (7,3))
ax.bar (x,y1, color = "yellow")
ax.plot (x,y2, color = "red")

fig.show()
```

```python
#adding Stanadard deviation
y2_sd = np.std (y2)
y1_sd = np.std (y1)

x = np.array(["chicken","duck","dog","rat","rabbit"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax =plt.subplots (figsize = (7,3))
ax.bar (x,y1, color = "yellow")
ax.plot (x,y2, color = "red")

ax.errorbar(x,y2, yerr= y2_sd, fmt="none")

fig.show()
```

```python
#adding Stanadard deviation and modifying elements of the error bar
y2_sd = np.std (y2)
y1_sd = np.std (y1)

x = np.array(["chicken","duck","dog","rat","rabbit"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax =plt.subplots (figsize = (7,3))
ax.bar (x,y1, color = "yellow")
ax.plot (x,y2, color = "red")

ax.errorbar(x,y2, yerr= y2_sd, fmt="none",
            ecolor = "indigo",
            elinewidth= 4,
            capsize= 6,
            capthick= 4)

fig.show()
```

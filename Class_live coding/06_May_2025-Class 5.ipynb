```python
# making Vis with SEABORN
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#load data set
tips = sns.load_dataset ("tips")
print (tips)
```

```python
sns.lineplot(data = tips,
             x= "total_bill",
             y = "tip")
```

```python
sns.set_style ("whitegrid")
sns.lineplot(data = tips,
             x= "total_bill",
             y = "tip")
```

```python
fig = plt.subplots (figsize = (10,3))

tipgraph = sns.lineplot(data=tips,
                        x = 'total_bill',
                        y = 'tip')
tipgraph.set(title = 'tips vs total bill',
             xlabel = 'total bill($)',
             ylabel ='tip amount')
```

```python
fig = plt.subplots (figsize = (10,3))

tipgraph = sns.lineplot(data=tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'purple',
                        linestyle = '--',
                        linewidth = 3,
                        marker ='o',
                        markerfacecolor = 'hotpink')
tipgraph.set(title = 'tips vs total bill',
             xlabel = 'total bill($)',
             ylabel ='tip amount')
```

```python
# making Vis with Plotly
import plotly.graph_objects as go 
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles,venn2_unweighted
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#load data
x2 = np.array(["chicken","duck","dog","rat","rabbit"])
y2 = np.array([110,180,240,99,220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x2,y=y2))
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x2,y=y2))

graph.update_layout (
    title = 'Animal Scores' ,
    xaxis_title = "Animals",
    yaxis_title = "Scores"
)
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x2,y=y2, mode = 'markers',
    marker = dict(
        size = 15,
        color = 'blue',
        opacity = 1,
        line = dict (width = 5, color = 'black')
    )))

graph.update_layout (
    title = 'Interactive Animal Scores' ,
    xaxis_title = "Animals",
    yaxis_title = "Scores",
    width = 500, height = 500
)
```

```python
df = pd.read_csv ("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",on_bad_lines= 'skip')
df
```

```python
text = " ". join (each  for each in df.quote)

wordcloud = WordCloud (background_color= "white",
                       colormap= "inferno").generate(text)

fig, ax = plt.subplots (figsize = (7,3))
ax.imshow (wordcloud, interpolation= "bilinear")
ax.axis("off")
```

```python
A = set(["chicken","duck","turkey"])
B = set(["cow","fish","dog","chicken"])
```

```python
diagram = venn2_unweighted([A,B],
                        set_labels = ("wild","Farm"),
                        set_colors = ("blue","red"),
                        alpha = 0.5)
# shows count of unique items
```

```python
diagram = venn2_unweighted([A,B],
                        set_labels = ("wild","Farm"),
                        set_colors = ("blue","red"),
                        alpha = 0.5)

diagram.get_label_by_id ("10") .set_text ("\n".join(A - B)) # join and display words only in A
diagram.get_label_by_id ("11").set_text ("\n".join(A & B)) # join  and display words only show the word in both A and B
diagram.get_label_by_id ("01").set_text ("\n".join(B - A)) # join and display words only in B
```

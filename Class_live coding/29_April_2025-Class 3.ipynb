```python
#load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# generating data
np.random.seed (1)
x = np.arange (51) # 0-50
y1= np.random.randint (0,100,51) # random integers 0-100, every 50?
y2= np.random.randint (0,100,51)
```

```python
# line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x,y1)
ax.plot (x,y2)
```

```python
#adding a legend to the graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x, y1, label = "Car 1")
ax.plot (x, y2, label = "Car 2")
ax.legend (loc = "upper left") #defining location
fig.show()
# question: how do i remove the box around the legend
```

```python
#changing attributes of legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x, y1, label = "Car 1")
ax.plot (x, y2, label = "Car 2")
ax.legend (loc = "upper left",
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False
           ) 
fig.show()
```

```python
# moving legend outside of plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x, y1, label = "Car 1")
ax.plot (x, y2, label = "Car 2")
ax.legend (loc = "upper left",
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,1) #horizontal at 100% and 100% on vertical axis
           ) 
fig.show()
```

```python
# moving legend outside of plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot (x, y1, label = "Car 1")
ax.plot (x, y2, label = "Car 2")
ax.legend (loc = "upper left", # refering to the legend its self, when used in conjuction with bbox
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,0.6) #horizontal at 100% and 60% on vertical axis
           ) 
fig.show()
```

```python
# adding annotation  with text() fxn
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left", # refering to the legend its self, when used in conjuction with bbox
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,0.6) #horizontal at 100% and 60% on vertical axis
           ) 
ax.text (10,105, "this value is important") # position from the beginning of text on x-axis and end on y-axis
fig.show()
```

```python
# adding annotation  with text() fxn
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left", # refering to the legend its self, when used in conjuction with bbox
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,0.6) #horizontal at 100% and 60% on vertical axis
           ) 
ax.text (10,105, "This value is important", # position from the beginning of text on x-axis and end on y-axis
         ha = "center",
         color = "purple",
         size = 20) 
fig.show()
```

```python
# annotating with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left", # refering to the legend its self, when used in conjuction with bbox
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,0.6) #horizontal at 100% and 60% on vertical axis
           ) 
ax.annotate ("this is important", xy=(17,95), xytext= (30,94),
             arrowprops = dict (facecolor = "pink",shrink = 0.10)) # shrink = 5% gap from point to arrow tip
fig.show()
```

```python
# changing arrow style
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left", # refering to the legend its self, when used in conjuction with bbox
           frameon = False, #removing frame 
           fontsize = 9.5,
           ncol = 1, #specify number of columns
           shadow = False,
           bbox_to_anchor = (1,0.6) #horizontal at 100% and 60% on vertical axis
           ) 
ax.annotate ("this is important", xy=(17,95), xytext= (30,94),
             arrowprops = dict (arrowstyle = "wedge", facecolor = "hotpink")) 
fig.show()
```

```python
# removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
ax.yaxis.set_major_locator(plt.NullLocator()) # remove number and tick marks
ax.xaxis.set_major_formatter(plt.NullFormatter()) # remove number but not tick marks
fig.show()
```

```python
# removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_major_formatter(plt.NullFormatter())
fig.show()
```

```python
# changing scale
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
ax.yaxis.set_major_locator (plt.MaxNLocator(4))
ax.xaxis.set_major_locator (plt.MaxNLocator(3))
fig.show()
```

```python
# changing scale intervals i.e. more tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
ax.yaxis.set_major_locator (plt.MultipleLocator(10))
ax.xaxis.set_major_locator (plt.MultipleLocator(3))
fig.show()
```

```python
# rotating tick mark numbers
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
plt.yticks (rotation = 45, ha = "right")
plt.xticks (rotation = 60, ha = "left")
fig.show()
```

```python
# rotating tick mark numbers
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 
plt.yticks (rotation = 45, ha = "left") # direction of tilt
plt.xticks (rotation = 45, ha = "right") # direction of tilt
fig.show()
```

```python
# adding Axis Title
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter (x, y1, label = "Car 1")
ax.scatter (x, y2, label = "Car 2")
ax.legend (loc = "upper left") 

font1 = { "family":"serif", "color":"indigo"}
plt.xlabel ("shinny new X-Axis", fontsize = 16, fontdict = font1)
plt.ylabel ("shinny new Y-Axis", fontsize = 16, fontdict = font1)
fig.show()
```

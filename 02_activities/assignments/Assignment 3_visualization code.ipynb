```python
import pandas as pd

Flood_Toronto = pd.read_csv("/Users/Ezynne/Documents/DSI/visualization/02_activities/assignments/Basement Flooding Event Data_Toronto_python raw data_Assignment 3_CSV.csv")
print(Flood_Toronto)
```

```python
# line plot
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(Flood_Toronto["Wards"],Flood_Toronto["BF"],label = "Total Basement Flooding (BF) (2005-2023)",linestyle = "--")
ax.plot(Flood_Toronto["Wards"],Flood_Toronto["BFPSP"],label = "Total Basement Floor Water Seepage(BFPSP) (2013-2023)")
ax.legend (loc = "upper right",
           frameon = False, #removing frame 
           fontsize = 9,
           ncol = 1, #specify number of columns
           shadow = False)
ax.text (10,8000, "Comparison of total  BF and total BFPSP by ward in Toronto ",fontweight = "bold")
ax.set_xticks(np.arange(1,47,1), labels=[str(i) for i in range(1, 47)])
font1 = { "family":"serif", "color":"Black"}
plt.xlabel ("Ward number", fontsize = 12, fontdict = font1, fontweight = "bold")
plt.ylabel ("Total number of flood events", fontsize = 12, fontdict = font1, fontweight = "bold")
fig.show()
```
